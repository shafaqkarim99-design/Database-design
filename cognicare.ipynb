```python
# Load the necessary CSV files using the correct paths
order_df = pd.read_csv('order.csv')
product_df = pd.read_csv('product.csv')
category_df = pd.read_csv('category.csv')

# Merge orders with products to get category information
order_category_df = pd.merge(order_df, product_df[['product_id', 'category_id']], on='product_id', how='inner')
order_category_df = pd.merge(order_category_df, category_df, on='category_id', how='inner')

# Group by category and count the orders
category_order_count = order_category_df.groupby('category_name').size().reset_index(name='order_count')

# Plotting order count by category
plt.figure(figsize=(10, 6))
sns.barplot(data=category_order_count, x='category_name', y='order_count', palette='Set2')
plt.title('Order Count by Category')
plt.xlabel('Category')
plt.ylabel('Order Count')
plt.xticks(rotation=45)
plt.show()

```

```python
# Merge the order_category_df with the product_df to get the product names
order_category_df = pd.merge(order_category_df, product_df[['product_id', 'product_name']], on='product_id', how='inner')

# Group by product name and count the orders
top_products = order_category_df.groupby('product_name').size().reset_index(name='order_count')

# Sort products by order count and get the top 10
top_products = top_products.sort_values('order_count', ascending=False).head(10)

# Plotting top 10 most ordered products
plt.figure(figsize=(10, 6))
sns.barplot(data=top_products, x='order_count', y='product_name', palette='coolwarm')
plt.title('Top 10 Most Ordered Products')
plt.xlabel('Order Count')
plt.ylabel('Product Name')
plt.show()
```
